```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd 
import seaborn as sns
from sklearn.model_selection import train_test_split
sns.set_palette('husl')
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

```python
#load the dataset
dataset = pd.read_csv('Iris_dataset.csv')
#dataset.describe()
```

```python
label_mapping = {
    'Iris-setosa' : 1,
    'Iris-versicolor' : 2,
    'Iris-virginica' : 3
}
```

```python
dataset_X = dataset.drop(['Species'], axis=1).values
dataset_Y = dataset.Species.replace(label_mapping).values.reshape(dataset.shape[0], 1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(dataset_X, dataset_Y, test_size=0.3, shuffle = True, random_state = 123)
```

```python
#question:
#Why do we need a dataset for train and a dataset for test? 
#Why can we use the real data for training?
#Where does the train data come from?
```

```python
#training the model using dataset
from sklearn import svm
regr = svm.SVR()
regr.fit(X_train, y_train.ravel())
#we have to ravel() the y data because it's in a weird format
#ravel() just flattens it into a nice 1D array
```

```python
#making the trained model predict output results from the dataset input
prediction = regr.predict(X_test)
print(prediction)
```

```python
#comparing prediction and real data
from sklearn.metrics import mean_squared_error
print('the mean squared error ($.2f) is:')
print(mean_squared_error(y_test, prediction))
from sklearn.metrics import r2_score
print('the coefficient of determination ($.2f) is:')
print(r2_score(y_test, prediction))
```
